```python
#チームごとに切り取り
left = 4
up = 8
width = 352
height = 196
x_margin = 17
y_margin = 8

teams = []
for row in range(4):
    if row != 0:
        up += height + y_margin
    for col in range(5):
        if col != 0:
            left += width + x_margin
        else:
            left = 4
        team = im_crop.crop((left, up, left+width, up+height))
        teams.append(team)

teams[11]
```
